```python
#Filtro por columnas (Usamos solo las columnas de variables cuantitativas)
filtro_columnas_innecesarias=data2_Micro_Retailer.iloc[ : , [5,6,13,14,16,21,22,31,42,49,51,59,29,33,78,90]]
filtro_columnas_innecesarias
```

```python
#Mtodo aplicando cuartiles Encuentro cuartules de 0.25 y 0.75
y=filtro_columnas_innecesarias
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del dataframe 
outliers_iqr_Micro_retailer=y[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr_Micro_retailer
```

```python
#Obtenemos datos limpios del dataframe
data3_iqr_Micro_Retailer=y[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr_Micro_Retailer
```

```python
#Obtenemos info de las columnas para corroborar que estemos ocupando los datos apropiados
data3_iqr_Micro_Retailer.info()
```

```python
#corroboramos valores nulos
valores_nulos=y.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atpicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=y.copy()
data4_iqr=y.fillna(round(data4_iqr.mean(),1))
data4_iqr
```

```python
#Observamos variables dentro del dataframe
data4_iqr
```

```python
#analizamos caso 1 de correlacin entre variable de nmero de clientes en la tienda respecto al nmero de estantes en la tienda, el nmero de refrigeradores y el nmero de empleados permanentes
Vars_Indep= data4_iqr[['268_number_fridges','104_how_many_shelves_does_the_micro_retailer_have', '2_current_permanent_employees']]
Var_Dep= data4_iqr['97_number_of_customers_in_store']
```

```python
#se define modelo como la funcin de regresin lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la funcin relacionada al modelo 
type(model)
```

```python
#ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R determinado
model.score(Vars_Indep,Var_Dep)
```

```python
#Declaramos las variables dependientes e independientes para la regresin logstica
Vars_Indep= data4_iqr[['268_number_fridges','104_how_many_shelves_does_the_micro_retailer_have', '2_current_permanent_employees']]
Var_Dep= data4_iqr['97_number_of_customers_in_store']
```

```python
#Redefinimos las variables
X= Vars_Indep
y= Var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =None)
```

```python
#Se escalan todos los datos
escalar= StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, utilizaremos fit_transform
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamos una prediccin
y_pred= algoritmo.predict(X_test)
y_pred 
```

```python
#Verifico la matriz de Confusin
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusin:')
print(matriz)
```

```python
#Calculo la precisin del modelo 
from sklearn.metrics import precision_score
precision= precision_score(y_test, y_pred, average="micro", pos_label="yes")
print('Precision del modelo:')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score
exactitud= accuracy_score(y_test, y_pred)
print('Exactitud del modelo')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo 
from sklearn.metrics import recall_score
sensibilidad=recall_score(y_test, y_pred, average='micro', pos_label="yes")
print('Sensibilidad del modelo')
print(sensibilidad)
```
